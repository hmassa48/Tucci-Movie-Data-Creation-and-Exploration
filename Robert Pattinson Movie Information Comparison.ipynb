### Get Data With Beautiful Soup

#### Import necessary libraries

```python
from bs4 import BeautifulSoup as bs
import requests
```

#### Collect the Info Box Information

```python
r = requests.get('https://en.wikipedia.org/wiki/Vanity_Fair_(2004_film)')

#convert to beautiful soup object
soup = bs(r.content)

#print html
contents = soup.prettify()
print(contents)
```

```python
#get just the info box of the movie

info_box = soup.find(class_='infobox vevent')
info_rows = info_box.find_all('tr')
for row in info_rows:
    print(row.prettify())
```

#### Looking at this Data we need a bit more data cleaning

```python
## Simple String Replacement \xa0 becomes a space
def get_content_value(row_data):
    if row.find('li'):
        return [li.get_text(" ", strip = True).replace("\xa0" , " " )for li in row.find_all('li')]
    else:
        return row_data.get_text(" ", strip = True).replace("\xa0" , " " )


movie_info = {}



for index, row in enumerate(info_rows):
    if index == 0:
        movie_info['title'] = row.find('th').get_text(" ", strip = True)
    elif index ==1:
        continue 
    else:
        content_key = row.find('th').get_text(" ", strip = True)
        content_value = get_content_value(row.find('td'))
        movie_info[content_key] = content_value
movie_info 
```

## Get All Robert Pattinson Movies

Now that we have set up a system to retrieve the information from one of Robert's films. We have to set up a pipeline to scrape all of his movie information. 

```python
r = requests.get('https://en.wikipedia.org/wiki/Robert_Pattinson')

#Convert to beautiful soup object
soup = bs(r.content)

#print html
contents = soup.prettify()
print(contents)
```

```python
## Get only the Film Table from the HTML Page

table = soup.find_all('table', class_="wikitable")[1]   # Only use the first table
```

```python
#Only the Links are italicized, select the links of his movies
movies = table.select("i a")
```

As a note we are only selecting movies with a link element. This is for data cleaning. In this case the movie Fear & Shame did not have a Wikipedia page. So this could be further explored in another iteration. 

```python
#example to see the link
movies[0]['href']
```

```python
## Create a Function to Take in a URL 
def get_info_box(url):
    
    r = requests.get(url)
    
    #Convert to beautiful soup object
    soup = bs(r.content)
    
    #Print out the HTML 
    contents = soup.prettify()
    
    info_box = soup.find(class_='infobox vevent')
    info_rows = info_box.find_all('tr')
    
    
    movie_info = {}

    for index, row in enumerate(info_rows):
        if index == 0:
            movie_info['title'] = row.find('th').get_text(" ", strip = True)
        elif index ==1:
            continue 
        else:
            content_key = row.find('th').get_text(" ", strip = True)
            content_value = get_content_value(row.find('td'))
            movie_info[content_key] = content_value
            
    return movie_info
```

```python

movie_info_list = []
base_path = 'https://en.wikipedia.org/'
for index, movie in enumerate(movies) :
    try:
        relative_path = movie['href']
        full_path = base_path + relative_path
        title = movie['title']
        
        movie_info_list.append(get_info_box(full_path))
        
    except Exception as e:
        print(movie.get_text())
        print(e)
        print()
    
```

### Save / Reload Movie Data

```python
import json 

def save_data(title,data):
    with open(title, 'w', encoding = 'utf-8') as f:
        json.dump(data,f,ensure_ascii = False, indent=2)
```

```python
import json 

def load_data(title):
    with open(title,encoding = 'utf-8') as f:
        return json.load(f)
```
